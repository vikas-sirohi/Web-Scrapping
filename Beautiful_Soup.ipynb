```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

# Scrapping Page 1<br>
 We will use this code to generalize, and scrap from multiple pages.

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1', headers = headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
companies = soup.find_all('div', class_ = "companyCardWrapper")
```

```python
# On page 1
#Features
name = []
rating = []
reviews_K = []
highly_rated = []
ctype = []
hq = []
no_of_employees= []
desc = []



for company in companies:
    name.append(company.find('h2').text.strip())
    rating.append(float(company.find('span', class_ = "companyCardWrapper__companyRatingValue").text.strip()))
    reviews_K.append(float(company.find('span', class_ ="companyCardWrapper__ActionCount").text.strip()[:-1]))
    highly_rated.append(company.find('span', class_ = "companyCardWrapper__ratingValues").text.strip())
    hq.append(company.find('span', class_="companyCardWrapper__interLinking").text.strip().split("|")[-1].split("+")[0])
    ctype.append(company.find('span', class_="companyCardWrapper__interLinking").text.strip().split("|")[0])
    no_of_employees.append(company.find('span', class_="companyCardWrapper__interLinking").text.strip().split("|")[1][:-10])

    company_url = "https://www.ambitionbox.com/overview/{}-overview".format(company.find('h2').text.strip())
    webpage_desc = requests.get(company_url, headers = headers).text
    soup_desc = BeautifulSoup(webpage_desc, 'lxml')
    #desc.append(soup_desc.find('div', class_="description mt-12").text)
    try:
      desc.append(soup_desc.find('div', class_="description mt-12").text)
    except:
      desc.append(np.nan)
```

```python
# Checking how much data is scraped
dictionary_pg1 ={'name': name, 'rating': rating, 'ctype': ctype, 'highly_rated': highly_rated, 'hq':hq, 'no_of_employees': no_of_employees,'reviews_K':reviews_K ,'desc': desc}
#df_pg1 = pd.DataFrame(dictionary_pg1)
vals = [len(val) for val in dictionary_pg1.values()]
print(vals)
```

So, There are 20 company per page. Let's convert it to a pandas dataframe.

```python
page_1_dict = {'name': name, 'rating': rating, 'ctype': ctype, 'highly_rated': highly_rated, 'hq':hq,'reviews_K':reviews_K, 'no_of_employees': no_of_employees,'desc': desc}

df_page_1 = pd.DataFrame(page_1_dict)
df_page_1 = df_page_1[["name", "rating","ctype", "highly_rated","hq", "reviews_K", "no_of_employees",  "desc" ]]
```

# Generalization <br>
The website will not allow us to scrape buch of pages in one go, so we will scrap say 10 pages per cycle and merge the data frames. <br>
Change the argument of the range to scrap pages, I have provided an example here.


```python
name = []
rating = []
reviews_K = []
highly_rated = []
ctype = []
hq = []
no_of_employees= []
desc = []
for j in range(2, 13):
  url = 'https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j)
  headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
  webpage = requests.get(url, headers = headers).text

  soup = BeautifulSoup(webpage, 'lxml')
  companies = soup.find_all('div', class_ = "companyCardWrapper")

  for company in companies:
      name.append(company.find('h2').text.strip())
      rating.append(float(company.find('span', class_ = "companyCardWrapper__companyRatingValue").text.strip()))
      reviews_K.append(float(company.find('span', class_ ="companyCardWrapper__ActionCount").text.strip()[:-1]))

      try:
        highly_rated.append(company.find('span', class_ = "companyCardWrapper__ratingValues").text.strip())
      except:
        highly_rated.append(np.nan)

      hq.append(company.find('span', class_="companyCardWrapper__interLinking").text.strip().split("|")[-1].split("+")[0])
      ctype.append(company.find('span', class_="companyCardWrapper__interLinking").text.strip().split("|")[0])

      try:
        no_of_employees.append(company.find('span', class_="companyCardWrapper__interLinking").text.strip().split("|")[1][:-10])
      except:
         no_of_employees.append(np.nan)

      company_url = "https://www.ambitionbox.com/overview/{}-overview".format(company.find('h2').text.strip())
      webpage_desc = requests.get(company_url, headers = headers).text
      soup_desc = BeautifulSoup(webpage_desc, 'lxml')
      try:
        desc.append(soup_desc.find('div', class_="description mt-12").text)
      except:
        desc.append(np.nan)
```

Simply creating pandas data frame from list, by first converting to dictionary

```python
dictionary ={'name': name, 'rating': rating, 'ctype': ctype, 'highly_rated': highly_rated, 'hq':hq,'reviews_K':reviews_K, 'no_of_employees': no_of_employees,'desc': desc}

df = pd.DataFrame(dictionary)
```

```python
df = df[["name", "rating","ctype", "highly_rated","hq", "reviews_K", "no_of_employees",  "desc" ]]
```

```python
df
```

Now, lets merge two data frame, from page 1 to page 12.

```python
merge = [df_page_1, df]
```

```python
updated = pd.concat(merge, ignore_index = True)
```
